```python
# Random Forest
import pandas as pd

#connect to google drive
# from google.colab import drive
# drive.mount('/content/drive')

# read the data and store data in DataFrame titled melbourne_data
path = "drive/MyDrive/Colab Notebooks/dataset/"
#df=pd.read_excel(path+'Telco_customer_churns.xlsx')
#melbourne_data = pd.read_csv('drive/My Drive/Colab Notebooks/kaggle_intro/melb_data.csv')
melbourne_data = pd.read_csv(path+'melb_data.csv')
```

```python
# Filter rows with missing values
melbourne_data = melbourne_data.dropna(axis=0)
# Choose target and features
y = melbourne_data.Price
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'BuildingArea', 
                        'YearBuilt', 'Lattitude', 'Longtitude']
X = melbourne_data[melbourne_features]
```

```python
X.describe()
```

```python
X.columns
```

```python
from sklearn.model_selection import train_test_split

# split data into training and validation data, for both features and target
# The split is based on a random number generator. Supplying a numeric value to
# the random_state argument guarantees we get the same split every time we
# run this script.
train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 0)
```

```python
#build a random forest model similarly to how we built a decision tree in scikit-learn

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor(random_state=1)

```

```python
forest_model.fit(train_X, train_y)
melb_preds = forest_model.predict(val_X)
print(mean_absolute_error(val_y, melb_preds))
```
